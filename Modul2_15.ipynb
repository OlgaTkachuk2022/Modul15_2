```python
import pandas as pd

# create a dictionary with data
data = {'month': ['Jan', 'Feb', 'Mar', 'Apr'],
        'pricePLN': [100, 200, 150, 300]}

# create a DataFrame object
df = pd.DataFrame(data)

# set the index to month
df.set_index('month', inplace=True)

# add a new column priceUSD
df['priceUSD'] = df['pricePLN'] / 4

# print the DataFrame
print(df)
```

```python
import matplotlib.pyplot as plt

# plot the DataFrame as a red dashed line graph
plt.plot(df['priceUSD'], 'r--')

# add labels and title
plt.xlabel('Month')
plt.ylabel('Price in USD')
plt.title('Price in USD by Month')

# display the graph
plt.show()
```
